```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
import pgeocode
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_auc_score
```

```python
data = pd.read_csv("Bank_Personal_Loan_Modelling.csv")
data.head()
```

```python
#adding data into DF
df = pd.DataFrame(data) 
# dropping ID coloumn
df.drop(['ID'] , axis = 1 , inplace = True) 

df.head()
```

```python
#Cheching for duplicated values :

print(len(df[df.duplicated()]))
```

```python
df.describe().T
```

```python
#Correcting noises :

#Absolute of this column
df['Experience'] = df['Experience'].abs()
```

```python
#Correcting "CCAvg" column :
# getting anuual value now by multiply with 12

df['CCAvg'] = df['CCAvg'].astype(float)*12
df.head()
```

```python
#Seperating discrete and continuous columns :

discrete = ['Family' , 'Education' ,'Personal Loan' , 'Securities Account' , 'CD Account' , 'Online' , 'CreditCard']
continuous = ['Age' , 'Experience', 'Income', 'CCAvg' , 'Mortgage']
```

```python
#Values of discrete columns :

for i in discrete :
    print(f' {i} : {df[i].unique()}')
```

```python
#Personal Loan is the column that i want to predict
x=df.drop('Personal Loan', axis='columns') #data
y=df['Personal Loan'] # label
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3) 
```

```python
forest=RandomForestClassifier()

forest.fit(x_train ,y_train)
```

```python
train_acc_forest= forest.score(x_train , y_train)
print("The Percentage of Random Forest model train accuracy is: ",train_acc_forest*100)
```

```python
y_prediction_forest = forest.predict(x_test)
```

```python
train_acc_forest= forest.score(x_train , y_train)
print("The Percentage of Random Forest model train accuracy is: ",train_acc_forest*100)
```

```python
test_acc_forest=accuracy_score(y_test ,y_prediction_forest )
print("The Percentage of Random Forest model test accuracy is: ",test_acc_forest*100)
```

```python
#Exctracting a dataframe that includes practical geographical information from 'ZIP Code' column using pgeocode library :

zip_codes = df["ZIP Code"].values.astype('str').tolist()
nomi = pgeocode.Nominatim('us')
us_code = nomi.query_postal_code(zip_codes)
us_code
```

```python
#Adding the 'place_name' column as 'City' to our dataframe :

us_code_renamed = us_code.rename(columns={'place_name': 'City'}) #Rename the column
df = pd.concat([df, us_code_renamed.City], axis=1 ) #Concat it to dataframe
df
```

```python
df.isna().sum()
#Checking for unknown zip codes :

df[df.City.isna()]['ZIP Code'].unique()
```

```python
df.loc[df['ZIP Code'] == 92717 ,'City'] = 'Irvine'
df.loc[df['ZIP Code'] == 92634 ,'City'] = 'Fullerton'
df.loc[df['ZIP Code'] == 92709 ,'City'] = 'Irvine'
```

```python
df.dropna(axis = 0 , inplace = True) #Drop the rows with unknown zip code
df.reset_index(inplace  = True , drop = True)
df
```
